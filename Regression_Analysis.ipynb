```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

Used to predict a numeric value (e.g., sales, price, demand) based on input features.
Types:
Linear Regression
Polynomial Regression
Logistic Regression (used for classification)
Use Case:
Predict customer lifetime value, product demand forecasting, or pricing optimization.

```python
#Advertising spend vs Sales
np.random.seed(42)
ad_spend = np.random.normal(1000, 200, 50)
sales = 30 + 0.05 * ad_spend + np.random.normal(0, 10, 50)

df = pd.DataFrame({'Ad_Spend': ad_spend, 'Sales': sales})

# Model
X = df[['Ad_Spend']]
y = df['Sales']
model = LinearRegression()
model.fit(X, y)

# Prediction
df['Predicted_Sales'] = model.predict(X)

# Plot
plt.figure(figsize=(8, 5))
plt.scatter(df['Ad_Spend'], df['Sales'], color='blue', label='Actual Sales')
plt.plot(df['Ad_Spend'], df['Predicted_Sales'], color='red', label='Regression Line')
plt.title("Linear Regression: Ad Spend vs Sales")
plt.xlabel("Ad Spend ($)")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()
```
