1.Importing all Required Librires for show case of Data Mining

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

 1. K-Means Clustering :- In data mining, K-means helps identify natural groupings or patterns within datasets, allowing for analysis and understanding of the data without predefined labels.

```python
# Simulated customer data for example
np.random.seed(42)
data = pd.DataFrame({
    "Annual Income (k$)": np.random.randint(15, 100, 100),
    "Spending Score (1-100)": np.random.randint(20, 100, 100)
})

# Elbow Method
wcss = []
X = data[["Annual Income (k$)", "Spending Score (1-100)"]]
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.grid(True)
plt.show()

# Apply KMeans
kmeans = KMeans(n_clusters=4, random_state=42)
data["Cluster"] = kmeans.fit_predict(X)

# Visualize
sns.scatterplot(data=data, x="Annual Income (k$)", y="Spending Score (1-100)", hue="Cluster", palette="Set2")
plt.title("Customer Segmentation")
plt.show()
```
