```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import zscore
```

 Anomaly Detection (Using Z-Score):-Z-score analysis is a statistical method used in data mining for anomaly detection, identifying data points that deviate significantly from the mean of a dataset

```python
#random transaction data
np.random.seed(42)
amounts = np.append(np.random.normal(100, 15, 45), [500, 550, 600, 700, 800])  

df = pd.DataFrame({'TransactionAmount': amounts})
df['Z_Score'] = zscore(df['TransactionAmount'])

# Identify anomalies
df['Anomaly'] = df['Z_Score'].apply(lambda x: 'Yes' if abs(x) > 2.5 else 'No')

# Plot
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x=range(len(df)), y='TransactionAmount', hue='Anomaly', palette={'Yes': 'red', 'No': 'green'})
plt.title("Anomaly Detection using Z-Score")
plt.xlabel("Transaction Index")
plt.ylabel("Transaction Amount")
plt.grid(True)
plt.show()
```
