```python
# Check for missing values in each column
null_counts = df.isnull().sum()
print(" Null values in each column:\n")
print(null_counts[null_counts > 0])
```

```python
print(df.dtypes)
```

```python
# Convert numeric columns to string before assigning 'unknown'
if df['Phone'].dtype != 'object':
    df['Phone'] = df['Phone'].astype(str)

# same for Zipcode if needed
if df['Zipcode'].dtype != 'object':
    df['Zipcode'] = df['Zipcode'].astype(str)

# Now fill missing values with correct data types
df.fillna({
    'Email': 'not_provided',
    'Phone': 'unknown',
    'Address': 'unknown',
    'Age': df['Age'].median(),
    'Feedback': 'No Feedback',
    'Ratings': df['Ratings'].mean(),
    'Customer_Segment': df['Customer_Segment'].mode()[0],
}, inplace=True)

print("Null values handled without type error.")
```

```python
# Check total duplicate rows
duplicate_rows = df.duplicated()
print(f" Total duplicate rows: {duplicate_rows.sum()}")
```

```python
# Remove duplicate rows
df = df.drop_duplicates()
print("Duplicate rows removed.")
```

```python
# Check for duplicates in important columns
key_columns = ['Transaction_ID', 'Customer_ID', 'Email', 'Phone']
duplicate_summary = {col: df[col].duplicated().sum() for col in key_columns}

print("Duplicate values per key column:")
for col, count in duplicate_summary.items():
    print(f"{col}: {count}")
```

```python
# Show cleaned DataFrame shape and preview
print("Final DataFrame shape:", df.shape)
df.head()
```

```python
df.to_csv('new_retail_data.csv')
```
