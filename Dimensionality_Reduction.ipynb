importing Required Libraries and module 

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

Used to reduce the number of features while retaining the essential patterns in the data.
Techniques:
Principal Component Analysis (PCA)
t-SNE (for visualization)
LDA (Linear Discriminant Analysis)

```python
iris = load_iris()
X = iris.data
y = iris.target

# PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(X)

# Prepare DataFrame
df_pca = pd.DataFrame(data=pca_result, columns=['PCA1', 'PCA2'])
df_pca['Target'] = y

# Plot
plt.figure(figsize=(8,6))
sns.scatterplot(data=df_pca, x='PCA1', y='PCA2', hue='Target', palette='tab10')
plt.title("PCA - Dimensionality Reduction on Iris Dataset")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)
plt.show()
```
